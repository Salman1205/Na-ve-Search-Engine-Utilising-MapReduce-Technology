```python
import pandas as pd
import string
import nltk
from nltk.corpus import stopwords


# file = 'enwiki-20170820.csv'
file = 'subset.csv'
df = pd.read_csv(file)
print(df.head(10))
```

```python
##Data Preprocessing
```

```python
def preprocess_text(text):
    punctuations = string.punctuation
    stop_words = set(stopwords.words('english'))
    
    text_without_punctuations = text.translate(str.maketrans('', '', punctuations))
    words = text_without_punctuations.split()
    filtered_words = [word.lower() for word in words if word.lower() not in stop_words]
    
    return filtered_words

df['SECTION_TEXT'] = df['SECTION_TEXT'].apply(preprocess_text)
print(df.head(10))
```

```python
#Vocabulary
all_words = []
for text in df['SECTION_TEXT']:
    all_words.extend(preprocess_text(str(text)))

# Create vocabulary of unique words
vocabulary = set(all_words)
print(vocabulary)
```

```python
# Assigning unique id to each word
word_to_id = {word: idx for idx, word in enumerate(vocabulary)}
print(word_to_id)
```

```python
# Term frequency
tf_dict = {}
for index, text in df['SECTION_TEXT'].head(2).items():
    words = preprocess_text(str(text))
    tf_dict[index] = {term: words.count(term) for term in vocabulary if words.count(term) > 0}

print("Term Frequency (TF) of top 2 documents (sparse representation):")
for doc_id, tf in tf_dict.items():
    print(f"Document {doc_id}: {tf}")
```
