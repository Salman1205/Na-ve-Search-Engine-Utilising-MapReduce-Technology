```python
import pandas as pd
import string
from nltk.corpus import stopwords
import math


# file = 'enwiki-20170820.csv'
file = 'subset.csv'
df = pd.read_csv(file)
print(df.head(10))
```

```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

def preprocess_text(text):
    punctuations = string.punctuation
    stop_words = set(stopwords.words('english'))
    lemmatizer = WordNetLemmatizer()
    
    text_without_punctuations = text.translate(str.maketrans('', '', punctuations))
    words = word_tokenize(text_without_punctuations)
    filtered_words = [lemmatizer.lemmatize(word.lower()) for word in words if word.lower() not in stop_words]
    
    return filtered_words

df['SECTION_TEXT'] = df['SECTION_TEXT'].apply(preprocess_text)
print(df.head(10))
```

```python
#Vocabulary
all_words = []
for text in df['SECTION_TEXT']:
    all_words.extend(preprocess_text(str(text)))

# Create vocabulary of unique words
vocabulary = set(all_words)
print(vocabulary)
```

```python
# Assigning unique id to each word
word_to_id = {word: idx for idx, word in enumerate(vocabulary)}
print(word_to_id)
```

```python
tf_dict = {}
for index, text in df['SECTION_TEXT'].head(5).items():
    words = preprocess_text(str(text))
    tf_dict[index] = {word_to_id[term]: words.count(term) for term in vocabulary if words.count(term) > 0}

print("Term Frequency (TF) of top 2 documents (sparse representation):")
for doc_id, tf in tf_dict.items():
    print(f"Document {doc_id}: {tf}")
```

```python
from collections import Counter

# Assuming `preprocess_text`, `word_to_id`, `vocabulary`, and `df` are defined elsewhere

# Initialize a Counter to hold the aggregated term frequencies
overall_tf = Counter()

# Iterate over the documents in the DataFrame
for index, text in df['SECTION_TEXT'].head(5).items():
    # Preprocess the text and get term frequencies for the document
    words = preprocess_text(str(text))
    tf = {word_to_id[term]: words.count(term) for term in vocabulary if words.count(term) > 0}
    
    # Update the overall term frequencies with the frequencies from this document
    overall_tf.update(tf)

# Print the overall term frequencies
print("Overall Term Frequency (TF) across all documents:")
print(overall_tf)
```

```python
# Calculate IDF for each word based on the top 2 documents
document_freq = {word: sum(1 for text in df['SECTION_TEXT'].head(5) if word in text) for word in vocabulary}
idf_dict = {word_to_id[word]: math.log(5 / (document_freq[word] + 1)) for word in vocabulary}

# Calculate TF-IDF weighted for each document
tf_idf_dict = {}
for doc_id, tf in tf_dict.items():
    tf_idf_dict[doc_id] = {word_id: tf_val * idf_dict[word_id] for word_id, tf_val in tf.items()}

# Convert TF-IDF weights to sparse vector representation
sparse_tf_idf_dict = {}
for doc_id, tf_idf in tf_idf_dict.items():
    sparse_tf_idf_dict[doc_id] = {word_id: tf_idf_val for word_id, tf_idf_val in tf_idf.items() if tf_idf_val != 0}

# Print TF-IDF weights of each word in each document in sparse vector representation
print("TF-IDF weights of each word in each document (Sparse Vector Representation):")
for doc_id, tf_idf in sparse_tf_idf_dict.items():
    print(f"Document {doc_id}: {{ {', '.join(f'{word}: {tf_idf[word]:.2f}' for word in tf_idf)} }}")

```

```python
# Query text
query_text = "the anarchist farm amateur dimensional occupies crime party violence"

# Preprocess the query text
query_words = preprocess_text(query_text)

# Calculate document frequency for the query words
query_document_freq = {word: sum(1 for text in df['SECTION_TEXT'] if word in text) for word in query_words}

# Calculate IDF for each word in the query
query_idf_dict = {word_to_id[word]: math.log(len(df) / (query_document_freq[word] + 1)) for word in query_words}

# Convert query IDF weights to vector representation
query_idf_vector = {word_id: idf for word_id, idf in query_idf_dict.items()}

print("Query IDF vector:")
for word_id, idf in query_idf_vector.items():
    print(f"Word ID {word_id}: {idf:.2f}")
```

```python
# Calculate dot product of query with each document
query_dot_product = {}

for doc_id, tf_idf in sparse_tf_idf_dict.items():
    dot_product = 0
    for word_id, idf_weight in query_idf_vector.items():
        if word_id in tf_idf:
            dot_product += idf_weight * tf_idf[word_id]
    query_dot_product[doc_id] = dot_product

print("Dot product of query with each document:")
for doc_id, dot_product in query_dot_product.items():
    print(f"Document {doc_id}: {dot_product:.2f}")
```
